Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion. 

```python
df3.plot.scatter(x='a',y='b', c = 'red', xlim=[-0.2,1.2], ylim=[-0.2,1.2], figsize=(12,3), s=30)
```

Create a histogram of the 'a' column.

```python
plt.style.use('ggplot')
df3['a'].hist()
```

figure out how to add more bins to it.

```python
df3['a'].hist(bins=30, alpha=.5)
```

Create a boxplot comparing the a and b columns.

```python
df3[['a','b']].plot.box()
```

Create a kde plot of the 'd' column

```python
df3['d'].plot.kde(lw=0.9)
```

Figure out how to increase the linewidth and make the linestyle dashed.

```python
df3['d'].plot.kde(lw=4, ls='--')
```

Create an area plot of all the columns for just the rows up to 30.

```python
df3[:31].plot.area(alpha=.4)
```

Can you figure out how to display the legend outside of the plot as shown below?

```python
df3[:31].plot.area(alpha=.4).legend(bbox_to_anchor=(1.0, 0.7))
```
